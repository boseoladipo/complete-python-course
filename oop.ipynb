```python
class Student:
    def __init__(self, name, school):
        self.name = name
        self.school = school
        self.marks = []

    def average(self):
        return sum(self.marks) / len(self.marks)


class WorkingStudent:
    def __init__(self, name, school, salary):
        super.__init__(name, school)
        self.salary = salary

    def average(self):
        return sum(self.marks)/ len(self.marks)

rolf = WorkingStudent('Rolf', "MIT", 15.50)
print(rolf.salary)
```
